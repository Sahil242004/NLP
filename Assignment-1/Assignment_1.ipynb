# Problem Statement : Perform tokenization (Whitespace, Punctuation-based, Treebank, Tweet, MWE) using NLTK library. Use porter stemmer and snowball stemmer for stemming. Use any technique for lemmatization.

Tokenization -  Process of breaking text into smaller tokens or chunks


 Tokens can be:
*   Words
*   sentences


*   Symbols
*  Phrases

Types of tokenization in this assignments

1.   Whitespace Tokenization

1.  Punctuation-based Tokenization

1.   Treebank Tokenization
2.   Tweet Tokenization
2.   Multi-Word Expression (MWE) Tokenization









**Step 1 - Install and import libraries**

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

**Step 2 -  Sample text input**

```python
text = "NLTK is a powerful library. I love working on NLP projects! Let's tokenize tweets like #AI @OpenAI "
print(text)
```

**Step 3 -** **Tokenization Implementation**

**Step 3.1 - Whitespace Tokenization**

```python
tokens_whitespace = text.split()
print("Whitespace Tokenization:")
print(tokens_whitespace)
```

**Step 3.2 - Punctuation-based Tokenization**

```python
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')
tokens_punctuation = tokenizer.tokenize(text)

print("Punctuation-based Tokenization:")
print(tokens_punctuation)
```

**Step 3.3 - Treebank Tokenization**

```python
from nltk.tokenize import TreebankWordTokenizer

treebank_tokenizer = TreebankWordTokenizer()
tokens_treebank = treebank_tokenizer.tokenize(text)

print("Treebank Tokenization:")
print(tokens_treebank)
```

**Step 3.4 -  Tweet Tokenization**

```python
from nltk.tokenize import TweetTokenizer

tweet_tokenizer = TweetTokenizer()
tokens_tweet = tweet_tokenizer.tokenize(text)

print("Tweet Tokenization:")
print(tokens_tweet)
```

**Step 3.5 - Multi-Word Expression (MWE) Tokenization**

```python
from nltk.tokenize import MWETokenizer

mwe_tokenizer = MWETokenizer([('machine', 'learning'), ('artificial', 'intelligence')], separator='_')

sample_text = "I am learning machine learning and artificial intelligence"
tokens_mwe = mwe_tokenizer.tokenize(sample_text.split())

print("MWE Tokenization:")
print(tokens_mwe)
```

**Step 4 - Stemming Implementation**

**Step 4.1 - Porter Stemmer**

```python
from nltk.stem import PorterStemmer

porter = PorterStemmer()
words = ["playing", "played", "plays", "studies", "running"]

print("Porter Stemmer:")
for word in words:
    print(word, "->", porter.stem(word))
```

**Step 4.2 - SnowBall Stemmer**

```python
from nltk.stem import SnowballStemmer

snowball = SnowballStemmer("english")

print("Snowball Stemmer:")
for word in words:
    print(word, "->", snowball.stem(word))
```

**Step 5 - Lemmatization Implementation**

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
words = ["running", "better", "cars", "studies"]

print("Lemmatization:")
for word in words:
    print(word, "->", lemmatizer.lemmatize(word))
```
